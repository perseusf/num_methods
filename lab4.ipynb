С точностью e = 10-3, 10-6, 10-9 найти приближённое значение корня уравнения, лежащее на интервале (0, 10). Для поиска корня использовать метод дихотомии и метод Ньютона.

#### Вариант 18

$$ \arcsin{\frac{x-5}{6}}-2e^{-x}-\frac{1}{2} = 0 $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0 #начало интервала
b = 10 #конец интервала

h = 1e-3

epsilon1 = 1e-3
epsilon2 = 1e-6
epsilon3 = 1e-9
```

```python
#изначальная функция f(x)
def function(x):
    return np.arcsin((x-5)/6) - 2*np.exp(-x) - 1/2
```

```python
#первая производная f(x)
def d1(x):
    return 1/np.sqrt(-x**2 + 10*x + 11) + 1/np.exp(x)
```

```python
#график функции f(x)
x = np.arange(a, b, epsilon1)
y = function(x)
```

```python
fig = plt.figure(figsize=(13,7)) 

#график f(x)
plt.plot(x, y, ':', color="black", label="график f(x)")

plt.title(f'f(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend(loc="upper left")
plt.show()
```

### Метод дихотомии

```python
def dichotomy(a, b, epsilon):
    c = 0.5*(a+b)
    if (b - a > epsilon) and abs(function(c)) > epsilon:
        if function(c) == 0:
            return c
        if function(a)*function(c) < 0:
            c = dichotomy(a, c, epsilon)
        if function(b)*function(c) < 0:
            c = dichotomy(c, b, epsilon)
    return c
```

```python
root1 = dichotomy(a, b, epsilon1)
print('x =', root1, "\nPrecision e =", epsilon1)
```

```python
root2 = dichotomy(a, b, epsilon2)
print('x =', root2, "\nPrecision e =", epsilon2)
```

```python
root3 = dichotomy(a, b, epsilon3)
print('x =', root3, "\nPrecision e =", epsilon3)
```

### Метод Ньютона

```python
def newton(x, epsilon):
    xn = list()
    xn.append(x)
    while True:
        xnext = x - (function(x) / d1(x))
        xn.append(xnext)
        if abs(xnext - x) <= epsilon and abs(function(xnext)) <= epsilon:
            break
        x = xnext
    return xnext
```

```python
root4 = newton((b-a)/2, epsilon1)
print('x =', root4, "\nPrecision e =", epsilon1)
```

```python
root5 = newton((b-a)/2, epsilon2)
print('x =', root5, "\nPrecision e =", epsilon2)
```

```python
root6 = newton((b-a)/2, epsilon3)
print('x =', root6, "\nPrecision e =", epsilon3)
```
